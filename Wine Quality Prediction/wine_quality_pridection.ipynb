```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
```

```python
# data collection
wine_dataset = pd.read_csv('WineQT.csv')
```

```python
#printing the first n rows of the data set
wine_dataset.head()
```

```python
#getting the statstical measures  of the data set 
wine_dataset.describe()
```

```python
wine_dataset.groupby('quality').mean()
```

```python
#observations when wine quality is good volatile acidity is less and when its bad volitile acidity is more
x = wine_dataset.drop('quality', axis=1)
print(x)
```

```python
y= wine_dataset['quality']
print(y)
```

```python
#label binirization
#now we will have value as zero or one 1 if it is>=7 or else 0
y = wine_dataset['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
print(y)
```

```python
#splitting the data into train and test 
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.2, stratify=y,random_state=2)
```

```python
print(x.shape, x_train.shape , x_test.shape)
```

```python
# model training using random forest classifier
model = RandomForestClassifier()
```

```python
model.fit(x_train, y_train)
```

```python

x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
print(test_data_accuracy)
```
